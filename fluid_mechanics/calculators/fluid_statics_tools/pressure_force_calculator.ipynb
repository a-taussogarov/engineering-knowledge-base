```python
import math
def calculate_height(p, rho, g=9.81): # Pa, kg/m3, m/s2
    if rho <= 0:
        return "error"
    h = p/(rho*g)
    return h
height_earth = calculate_height(101325, 1000)
print(f"height on earth is {height_earth:.2f} m")

def area(radius):
    A =  math.pi*(radius**2) # m2 | pi is not approximated to 3.14
    return A

cylinder_1 = area(2)
print(f"area of cylinder 1 is {cylinder_1:.2f} m2")
cylinder_2 = area(3)
print(f"area of cylinder 2 is {cylinder_2:.2f} m2")

def get_force(radius, pressure): # m, Pa
    s = area(radius) # m2
    f = s*pressure # N
    return f

pressure = 20000000 # Pa
rad = 0.1 # m

force_result = get_force(rad, pressure)
print(f"force is {force_result:.0f} N")

radii = [0.06, 0.1, 0.15]
for r in radii:
    f = get_force(r, 20000000)
    print(f"for r={r} force is {f:.0f} N")

depths = [1000, 1250, 1500] # m
rho_2 = 1100 # kg/m3
g = 9.81

print("depth (m) | pressure (Pa)")
print("-" * 25)

for h in depths:
    p_1 = rho_2*g*h
    print(f"{h:<3} m | {p_1:.0f} Pa")
    
print("-" * 25)
```
